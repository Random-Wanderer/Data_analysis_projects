```python
# Imports
import pandas as pd
import seaborn as sns
```

# 2017

```python
# Look at 6/12/2017
crypto_2017 = pd.read_csv('datasets/coinmarketcap_06122017.csv')
crypto_2017.head()
```

```python
# Info
crypto_2017.info()
```

```python
# Describe
crypto_2017.describe()
```

```python
# Check for duplicates
crypto_2017.duplicated().sum()
```

```python
# Number of different coins usage
counts_2017 = crypto_2017['name'].value_counts()
counts_2017
```

```python
# change index
new_index_2017 = crypto_2017.set_index('name')
new_index_2017
```

```python
# drop columns
numbers_2017 = new_index_2017.drop(labels=['Unnamed: 0', 'id', 'symbol', 'rank'], axis= 1)
numbers_2017
```

```python
# scatter
sns.scatterplot(data= numbers_2017['price_usd'])
```

```python
# Max USD
numbers_2017['price_usd'].max()
```

```python
# Find outlier for price
filter_outlier = numbers_2017[numbers_2017['price_usd'] == 393520]
filter_outlier
```

```python
# drop bit coin
drop_bit20 = numbers_2017.drop(labels= ['Bit20'], axis= 0)
drop_bit20
```

```python
sns.scatterplot(data= drop_bit20['price_usd'])
```

# 2018

```python
# Look at 6/12/2018
crypto_2018 = pd.read_csv('datasets/coinmarketcap_06012018.csv')
crypto_2018.head()
```

```python
# info for 2018
crypto_2018.info()
```

```python
# Describe 2018
crypto_2018.describe()
```

```python
# Duplicated
crypto_2018.duplicated().sum()
```

```python
# Crypto counts
crypto_2018['name'].value_counts()
```

```python
# Reset index
new_index_2018 = crypto_2018.set_index('name')
new_index_2018
```

```python
# Drop columns
numbers_2018 = new_index_2018.drop(labels=['Unnamed: 0', 'id', 'symbol', 'rank'], axis= 1)
numbers_2018
```

# Merged

```python
# 2017 and 2018 merged
merged = crypto_2017.merge(crypto_2018, how= 'right', on= 'name')
merged
```

```python
# Short table
short_table = merged[['name', 'last_updated_x', 'market_cap_usd_x', 'last_updated_y', 'market_cap_usd_y']]
short_table
```

# Visuals
